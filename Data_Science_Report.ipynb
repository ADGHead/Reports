## <u>Data Science</u>

<b>What is Data Science?</b>

- Data science can be defined as coding, mathematics and statistics in applied settings.
- Data science can also be defined as the analysis of diverse data.
- Data science can also be defined as the inclusive analysis of data.
- Basically, Data science is the gathering of unorganized data and then finding meaning and value to that data.

<u>Pathway Of Data Science</u>

According to Drew Conway, Data science can be defined as the intersection of these three fields:
1. Coding
2. Statistics
3. Domain Knowledge

<u>Venn Diagram</u>

As seen from the venn diagram, Data science is the intersection between Hacking or coding, Math and Statistics knowledge and Substantive Expertise of Domain Knowledge.

The pathways in data science stem from the intersections of the various fields in data science.
- The intersection between Hacking skills and Statistics is <b>Machine Learning</b>.
- The intersection between Statistics and Substantive Expertise is <b>Traditional Research</b>.
- The intersection between Hacking skills and Substantive Expertise is <b>Danger Zone!</b>.
